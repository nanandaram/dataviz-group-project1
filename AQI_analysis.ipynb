```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Load in AQI csv
AQI_df = pd.read_csv("Resources/AQI_data.csv")
AQI_df.head()
```

```python
# Load in Restrictions csv
lockdown_df = pd.read_csv("Resources/Lockdown_data_wiki.csv")
lockdown_df.head()
```

```python
# Combine the dataframes on the country column
combined_df = pd.merge(AQI_df,lockdown_df)
```

```python
# Print the header of the combined dataframe
combined_df.head()
```

```python
# Add column to track whether country was under lockdown each week
combined_df["lockdown"] = ""

# Determine whether each country was under lockdown each week 
for index, row in combined_df.iterrows():
    if (row["start_week"] > row["week"]) | (row["level"] == "None"):
        combined_df.loc[index, "lockdown"] = "No"
    else:
        combined_df.loc[index, "lockdown"] = "Yes"
```

```python
# Export to csv
combined_df.to_csv("foo.csv")
```
